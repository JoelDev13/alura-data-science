**Variables numéricas**


```
Escribe un programa en Python que:

Pida al usuario el precio de un producto

Pida al usuario la cantidad de unidades que quiere comprar

Calcule el total a pagar (precio × cantidad)

Muestre el resultado en pantalla
```




````python
precio = float(input("Ingrese el precio del producto: "))
cantidad = int(input("Ingrese la cantidad de unidades: "))

total = precio * cantidad

print("El total es:", total)
````

**String**

````python
saludo = " yolA BRO "

saludo = saludo.upper()
print(saludo)  # HOLA BRO

saludo = saludo.lower()
print(saludo)  # hola bro

saludo = saludo.strip()
print(saludo)  # hola bro (sin espacios al inicio o final)


saludo = saludo.strip().replace("y" , "h") #remplaza la y por la h
print(saludo)
````
